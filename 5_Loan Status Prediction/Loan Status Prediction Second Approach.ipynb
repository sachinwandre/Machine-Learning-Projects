```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

## Data Load: Loan Status Prediction into a dataframe

```python
loan_dataset  = pd.read_csv('Loan.csv')
loan_dataset .head()
```

```python
# dropping the missing values
loan_dataset = loan_dataset.dropna()
```

```python
# Checking for duplicates
loan_dataset.duplicated().sum()
```

```python
pd.set_option('future.no_silent_downcasting', True)
```

```python
# label encoding
loan_dataset.replace({"Loan_Status":{'N':0,'Y':1}},inplace=True)
```

```python
# printing the first 5 rows of the dataframe
loan_dataset.head()
```

```python
# Dependent column values
loan_dataset['Dependents'].value_counts()
```

```python
# replacing the value of 3+ to 4
loan_dataset = loan_dataset.replace(to_replace='3+', value=4)
```

```python
# dependent values
loan_dataset['Dependents'].value_counts()
```

```python
loan_dataset.info()
```

```python
loan_dataset['Education'] = loan_dataset['Education'].astype('category')
loan_dataset['Loan_Status'] = loan_dataset['Loan_Status'].astype('category')
```

```python
# convert categorical columns to numerical values
loan_dataset.replace({'Married':{'No':0,'Yes':1},'Gender':{'Male':1,'Female':0},'Self_Employed':{'No':0,'Yes':1},
                      'Property_Area':{'Rural':0,'Semiurban':1,'Urban':2},'Education':{'Graduate':1,'Not Graduate':0}},inplace=True)
```

```python
loan_dataset.head()
```

```python
# loan_dataset['TotalIncome'] = loan_dataset['ApplicantIncome'] + loan_dataset['CoapplicantIncome']
```

```python
# separating the data and label
X = loan_dataset.drop(columns=['Loan_ID','Loan_Status'],axis=1)
Y = loan_dataset['Loan_Status']
```

```python
# separating the data and label
X = loan_dataset.drop(columns=['Loan_ID','Loan_Status','ApplicantIncome','CoapplicantIncome'],axis=1)
Y = loan_dataset['Loan_Status']
```

```python
print(X.shape,Y.shape)
```

## Train Test Split

```python
X_train, X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Training the model

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the support Vector Macine model
classifier.fit(X_train,Y_train)
```

## Model Evaluation

```python
# accuracy score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuray = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data : ', training_data_accuray)
```
